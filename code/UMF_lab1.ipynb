# Схема уголок


Реализация схемы уголок для CFL1=0.6, 1, 1.01

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

CFL1 = 0.6 #число Куранта
L = 20    #размер области
T = 18    #время расчета
h = 0.5


Nx = 40  #число узлов по пространству

tau = CFL1 * h
Nt = 40

x = np.linspace(0, L, num = Nx)
 
y = np.zeros((Nt, Nx))

#задание НУ 
y[0, :] = np.sin(4 * np.pi * x / L)
t = 0

#цикл по времени и пр
for j in range(1, Nt):
    for k in range(2, Nx):
        y[j, k] = (tau*(y[j-1, k-1] - y[j-1, k]))/h + y[j-1, k]    #реализация схемы
        t = t+tau     #инкрементирование времени
    y[j,1] = y[j,Nx-1]    #ГУ

   
# Построение графика
plt.figure(figsize=(10, 6))
for v in range(0, Nt, 10):
    plt.plot(x, y[v, :], label=f"t={v}" )    #постороение графика решения в разные моменты времени
plt.xlabel('x')
plt.ylabel('y(x,t)')
plt.legend()
plt.title('Решение уравнения переноса схемой уголок с CFL1 = 0.6')
plt.grid(True) 
plt.show()       
print(t)
```

# Схема Л.-В. для уравнения переноса

Реализация схемы для CFL1=0.6, 1, 1.01

```python
import numpy as np
import matplotlib.pyplot as plt
import math

Nx = 40
Nt = 40
L = 20    #размер области
T = 18    #время расчета
h = 0.5
CFL1 = 0.6


tau = CFL1 * h


x = np.linspace(0, L, num = Nx+1)
y = np.zeros((Nt+1, Nx+1))

#задание НУ 
for i in range(0,Nx):
    y[0, i] = math.sin(4*np.pi*x[i]/L)

#цикл по времени и пр
for n in range(1, Nt-1):
    #ГУ:
    y[n,Nx] = ( (-1/(2*h))*(y[n-1,1] - y[n-1,Nx-1]) + (tau/(2*h*h))*(y[n-1,1] - 2*y[n-1,Nx] + y[n-1,Nx-1]) )*tau + y[n-1,Nx]
    y[n,1] = ( (-1/(2*h))*(y[n-1,2] - y[n-1,Nx]) + (tau/(2*h*h))*(y[n-1,2] - 2*y[n-1,1] + y[n-1,Nx]) )*tau + y[n-1,1]
        
    for m in range(2, Nx):
        #реализация схемы
        y[n,m] =  (tau/(2*h))*(-y[n-1,m+1] + y[n-1,m-1]) + (tau/(2*h*h))*(y[n-1,m+1] - 2*y[n-1,m] + y[n-1,m-1]) + y[n-1,m]
        

# Построение графика
plt.figure(figsize=(10, 6))
for n in range(0, Nt, 10):
    plt.plot(x, y[n, :], label=f"t={n}" )
plt.xlabel('x')
plt.ylabel('y(x,t)')
plt.legend()
plt.title('Решение схемой Лакса–Вендроффа CFL1 = 0.6')
        
plt.grid(True) 
plt.show()
```
